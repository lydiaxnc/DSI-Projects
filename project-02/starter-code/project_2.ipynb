## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. Explain your rationale for the way you choose to "impute" the missing data.

```python
# rename some of the columns 
bb.rename(columns = {'artist.inverted':'artist','date.entered':'date_entered','date.peaked':'date_peaked'}, inplace = True)
bb.columns
```

```python
# Break down the data inside of 'date_entered' and 'date_peaked' into year, month, and day
temp = pd.DatetimeIndex(bb['date_entered'])
bb['year_entered'] = temp.year
bb['month_entered'] = temp.month
bb['day_entered'] = temp.day

temp = pd.DatetimeIndex(bb['date_peaked'])
bb['year_peaked'] = temp.year
bb['month_peaked'] = temp.month
bb['day_peaked'] = temp.day

bb.columns
```

```python
# delete extra columns: 'date_entered' and 'date_peaked' 
bb = bb.drop(['date_peaked','date_entered'], axis=1)
```

```python
# check how many values are in 'year' column
print pd.value_counts(pd.Series('year'))
print pd.value_counts(pd.Series('year_entered'))
print pd.value_counts(pd.Series('year_peaked'))
```

```python
# all entries occurred in 2000, it's not necessary to keep those columns
bb = bb.drop(['year','year_entered','year_peaked'], axis=1)
bb.columns
```

```python
# rename per week's ranking columns
c_names1 = bb.columns[0:4].tolist()
c_names2 = [x+1 for x in range(76)]
c_names3 = bb.columns[80:].tolist()
c_names = c_names1 + c_names2 + c_names3
bb.columns = c_names
# check new column names
bb.columns
```

```python
# replacing '*' with nan
bb.replace(bb[76][0], np.nan, inplace=True)
# check ranking data
bb.iloc[:,4:80]
```

```python
# check datatypes
bb.dtypes
```

```python
# investigate why columns week66 - week76 is float
bb.iloc[:,69:80]
bb.iloc[:,69:80].notnull().values.any()

# drop those columns 
bb = bb.drop(bb.iloc[:,69:80], axis=1)
bb
```

```python
bb.iloc[:,4:69] = bb.iloc[:,4:69].applymap(float)
```

```python
bb.dtypes
```

```python
# convert time into numeric format that is easy to calculate and compare
# convert time into editable string first
def to_string(x):
    try:
        y = (str(x)).replace('AM','')
        z = (str(y)).replace(',','')
        return str(z)
    except:
        pass      
bb['time'] = bb['time'].apply(to_string)

# check datatype
bb.dtypes

# convert time into seconds
bb['time'] = [int(x[0])*60 + int(x[1:3]) for x in bb['time']]
bb.time
```

```python
# check columns and datatype
bb.columns
bb.dtypes
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

```python
# melt ranking columns into rows, there should be 317 * 76 rows after the melt
ranking = pd.melt(bb, id_vars=['track'], value_vars = c_names2, var_name = 'week', value_name = 'ranking')
ranking
```

```python
# delete extra columns: the weekly ranking columns
bb_m = bb.drop(bb.iloc[:,4:69], axis=1)
bb_m
```

```python
# combine new dataframe with previous one
bb_m = pd.merge(ranking, bb_m, how = 'left', on='track')
bb_m.info()
```

```python
# change datatypes for week and ranking columns
bb_m['week'] = bb_m[['week']].applymap(int)
bb_m.info()
```

```python
# eliminate duplicated data 
bb_m.drop_duplicates() 
bb_m.info()
```

```python
# delete rows that don't have ranking 
bb_m.dropna(inplace=True)
bb_m
bb_m.info()
```

```python
# get a basic sense of numeric columns
bb_m.describe()
```

```python
# sort data 
bb_m.sort_values(by=['ranking','month_entered','day_entered','month_peaked','day_peaked'], ascending=True, inplace=True)
bb_m
```

```python
# output the new dataframe into a csv file so that it can be read by Tableau 
bb.to_csv('bb_data.csv')
bb_m.to_csv('bbm_data.csv')
```
